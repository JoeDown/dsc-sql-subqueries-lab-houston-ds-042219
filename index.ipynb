## Select the Total Number of Orders for Each Product Name

Sort the results by the total number of items sold for that product.

````python
c.execute('''select a.code, a.name, a.count from (select p.productcode code, p.productname name, count(*) as count, o.ordernumber 
             from products p 
             join orderdetails o 
             using(productcode) group by productname) a order by a.count desc''').fetchall()

````

## Select the Product Name and the  Total Number of People Who Have Ordered Each Product

Sort the results in descending order.

````python
c.execute('''select a.product, a.count from (select p.productname as product, 
              d.ordernumber, c.customernumber, count(c.customernumber) as count
              from products p join orderdetails d using(productcode)
              join orders o using(orderNumber)
              join customers c using(customernumber) group by p.productname) a
              order by a.count desc''').fetchall()
````

## Select the Employee Number, Office Code, City (of the office), and Name (First and Last) of those Employees who Sold Products that Have Been Ordered by Less Then 20 people.

This problem is a bit tougher. To start, think about how you might break the problem up. Be sure that your results only list each employee once.

````python
c.execute('''                
                    SELECT 
                         p.productName product, count(p.productName) count, e.employeeNumber employee, 
                         e.officeCode code, f.city city,
                         e.lastName last
                     FROM products p
                         JOIN orderdetails d USING(productCode)
                         JOIN orders o USING(orderNumber)
                         JOIN customers c USING(customerNumber)
                         JOIN employees e ON (c.salesRepEmployeeNumber=e.employeeNumber)
                         JOIN offices f USING(officeCode) 
                     WHERE count(p.productName)<28
                     GROUP BY p.productName
                     
        ''').fetchall()
````

````python
c.execute('''
             
                
                SELECT A.product, A.count, A.employee, 
                           A.code, A.city, A.last 
                FROM A
                
                JOIN
                
                    (SELECT A.product product, A.count count, A.employee employee, 
                           A.cod code, A.city city, A.last last
                    
                        FROM
                             (SELECT 
                                 p.productName product, count(p.productName) count, e.employeeNumber employee, 
                                 e.officeCode code, f.city city,
                                 e.lastName last
                             FROM products p
                                 JOIN orderdetails d USING(productCode)
                                 JOIN orders o USING(orderNumber)
                                 JOIN customers c USING(customerNumber)
                                 JOIN employees e ON (c.salesRepEmployeeNumber=e.employeeNumber)
                                 JOIN offices f USING(officeCode)) A

                         GROUP BY product) B
                         
                ON(A.product = B.product)

                WHERE B.count < 27
                ORDER BY B.count DESC
                 
                 ''').fetchall()
````
